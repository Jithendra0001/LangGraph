```python
from langgraph.graph import StateGraph
from typing import TypedDict
```

```python
class ComplimentAgentState(TypedDict):
    message: str
```

```python
def ComplimentAgent(state: ComplimentAgentState) -> ComplimentAgentState:
    state["message"] = f"Hey {state['message']}, you are doing an amazing job"
    return state
```

```python
graph = StateGraph(ComplimentAgentState)
graph.add_node("Compliment_agent", ComplimentAgent)
graph.set_entry_point("Compliment_agent")
graph.set_finish_point("Compliment_agent")
```

```python
app = graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": "Jithendra"})
```
